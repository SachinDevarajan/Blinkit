```python
import pandas as pd
```

```python
cus_df = pd.read_csv("Raw Data\\blinkit_customers.csv")
ord_df = pd.read_csv("Raw Data\\blinkit_orders.csv")
ord_item_df = pd.read_csv("Raw Data\\blinkit_order_items.csv")
prod_df = pd.read_csv('Raw Data\\blinkit_products.csv')
feedback_df = pd.read_csv("Raw Data\\blinkit_customer_feedback.csv")
delivery_df = pd.read_csv("Raw Data\\blinkit_delivery_performance.csv")
inventory_df = pd.read_csv("Raw Data\\blinkit_inventory.csv")
inventory_new_df = pd.read_csv("Raw Data\\blinkit_inventoryNew.csv")
marketing_df = pd.read_csv("Raw Data\\blinkit_marketing_performance.csv")
category_icon_df = pd.read_excel("Raw Data\\Category_Icons.xlsx")
rating_icon_df = pd.read_excel("Raw Data\\Rating_Icon.xlsx")
```
